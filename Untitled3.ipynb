```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import datasets,layers,models
```

```python
(x_train,y_train),(x_test,y_test)=datasets.cifar10.load_data()
```

```python
x_train,x_test=x_train/255,x_test/255
class_names=['plane','car','bird','cat','deer','dog','frog','horse','ship','truck']
```

```python
plt.figure(figsize=(10,10))
for i in range(10):
  plt.subplot(5,5,i+1)
  plt.imshow(x_train[i])
 # plt.xticks([])
  # plt.yticks([])
  plt.axis('off')
  plt.grid(False)
  plt.title(class_names[int(y_train[i][0])])
plt.show()
```

```python
x_train=x_train[:20000]
y_train=y_train[:20000]
x_test=x_test[:4000]
y_test=y_test[:4000]

model=models.Sequential()
#model.add(models.Flatten(input_dim=10))
model.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Conv2D(64,(3,3),activation='relu'))
model.add(layers.MaxPooling2D((2,2)))
model.add(layers.Flatten())
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dense(10,activation='softmax'))
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test))
loss,accuracy=model.evaluate(x_test,y_test)
print(f'loss:{loss}')
print(f'accuracy:{accuracy}')
#model.save('image_classifier.model')

```

```python
plt.scatter(class_names,model.history.history['accuracy'])
plt.show()
from plotly.express import bar
figu=bar(x=class_names,y=model.history.history['accuracy'],color=class_names,color_continuous_scale='earth')
figu.show()
```
